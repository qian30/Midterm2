```python
import pandas as pd
data = pd.read_csv('data.csv')
```

```python
data = data[7:209]
data
```

```python
age = []
for d in data['Unnamed: 2']:
    age.append(d)
for i in range(len(age)):
    age[i] = age[i].replace("'", "").replace("+", "")
    age[i] = int(age[i])
```

```python
year = []
for i in data['Unnamed: 13']:
    year.append(i)
for i in range(len(age)):
    year[i] = int(year[i])
```

```python
youth = 0
young = 0
adults = 0
elderly = 0
for i in range(len(age)):
    if age[i] < 15:
        youth += year[i]
    elif age[i] < 25:
        young += year[i]
    elif age[i] < 65:
        adults += year[i]
    else:
        elderly += year[i]
total = sum(year)
print(youth)
print(young)
print(adults)
print(elderly)
```

```python
print("Youth: Total = {}, Proportion = {}%".format(youth, round(youth/total*100,2)))
print("Young adults: Total = {}, Proportion = {}%".format(young, round(young/total*100,2)))
print("Adults: Total = {}, Proportion = {}%".format(adults, round(adults/total*100,2)))
print("Elderly: Total = {}, Proportion = {}%".format(elderly, round(elderly/total*100,2)))
print("Total Population: {}".format(total))
```
